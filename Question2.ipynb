#Gathering Data from the dataset.

```python
%%bigquery --project datavids
SELECT *
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
LIMIT 10
```

#How many rows are in the dataset?


```python
%%bigquery --project datavids
SELECT 
COUNT(*) as total_rows
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
```

#What is the average trip total based on payment type?

```python
%%bigquery --project datavids
SELECT  payment_type,
AVG(trip_total) as avg_trip_total
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
Group by 	payment_type
```

# How many trips, that had a non-zero recorded time, were above the average time?

```python
%%bigquery --project datavids
SELECT  taxi_id,COUNT(*)as total_trips_abov_avgsec,
AVG(trip_seconds) as avg_trip_sec
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
Where trip_seconds is not null
Group by taxi_id
```
